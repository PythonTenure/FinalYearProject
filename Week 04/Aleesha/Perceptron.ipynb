# Perceptron Demo

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {'cgpa':[7.8 , 4.0 , 6.9, 8.9 , 2.2],
         'resume_score':[6.45 ,3.4,5.9,7.9,1.6],
         'placed':[1,0,1,1,0]
       }
```

```python
df = pd.DataFrame(data)
```

```python
print(df.shape)
df.head()
```

```python
X = df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_     #w1 , w2
```

```python
p.intercept_    #b
```

```python
from mlxtend.plotting import plot_decision_regions
```
